```python
#tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

text= "Natural language processing (NLP) is interesting "

tokens= word_tokenize(text)

print(tokens)
```

```python
#sentence tokenization
from nltk.tokenize import sent_tokenize

text= "python is easy. it is widely used in data science."

sentences= sent_tokenize(text)

print(sentences)
```

```python
#stop word removal
from nltk.corpus import stopwords
stop_words= set(stopwords.words('english'))

words= word_tokenize(text)

filtered_words= [w for w in words if w.lower() not in stop_words]

print(filtered_words)
print(stop_words)
```

```python
#stemming
#stemming reduces words to their root form

from nltk.stem import PorterStemmer

stemmer= PorterStemmer()

words= ['running', 'runs', 'ran', 'playing', 'gone', 'watched', 'studies']

for w in words:
  print(stemmer.stem(w))
```

```python
#lematization converts words to their dictionary form
from nltk.stem import WordNetLemmatizer

lemmatizer= WordNetLemmatizer()
print(lemmatizer.lemmatize('running'))
print(lemmatizer.lemmatize('best'))
print(lemmatizer.lemmatize('coding'))
print(lemmatizer.lemmatize('writing'))
print(lemmatizer.lemmatize('studies'))
print(lemmatizer.lemmatize('gone'))
```

```python
#parts of speech
#POS tagging identifies grammatical roles of words

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words= word_tokenize("Python is a powerful programming language")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
#word frequency count
from nltk.probability import FreqDist
text= "python is easy. it is widely used in data science."
words= word_tokenize(text)
fd= FreqDist(words)
print(fd.most_common())
```
